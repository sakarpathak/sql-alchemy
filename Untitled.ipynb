```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#getting the column names of measurement table
inspector=inspect(engine)
columns=inspector.get_columns('Measurement')
for c in columns:
    print(c["name"],c["type"])
```

```python
#visualizing a row of the measurement table
first_row=session.query(Measurement).first()
first_row.__dict__
```

```python
#Getting the latest date in the table
latest_date=session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(latest_date)
print(type(latest_date))
```

```python
# Calculate the date 1 year ago
year_ago=dt.date(2017,8,23)-dt.timedelta(days=365)
print(year_ago)
```

```python
precipitation_data=session.query(Measurement.date, Measurement.prcp)\
                    .filter(Measurement.date>="2016-08-23")\
                    .order_by(Measurement.date).all()
precipitation_data
```

```python
#Creating the DataFrame
df=pd.DataFrame(precipitation_data, columns=['date', "prcp"])
df.head()

#if we need certarin rows only we can pass the rows.
#df=pd.DataFrame(precipitation_data[100:200], columns=("date", "prcp"))

```

```python
#Setting date as index
df.set_index('date', inplace=True)
df.head()
```

```python
#Dropping the NANs
df.dropna(how="any", inplace=True)
df.head()
```

```python
# Design a query to show how many stations are available in this dataset?

#Total Number of Stations
Total_Stations=session.query(Measurement).group_by(Measurement.station).count()
print(Total_Stations)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

#Most active stations in descending order
most_active_desc=session.query(Measurement.station, func.count(Measurement.station))\
                .group_by(Measurement.station)\
                .order_by(func.count(Measurement.station).desc()).all()
for station in most_active_desc:
    print(station)
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?


#Details of the most active station
most_active_details=session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
            filter(Measurement.station=="USC00519281").all()
print(most_active_details)
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

temp_obs=session.query(Measurement.tobs).filter(Measurement.station=="USC00519281").filter(Measurement.date>=year_ago).all()
#print(temp_obs)
```

```python
#Creating the data frame
temp_df=pd.DataFrame(temp_obs, columns=['tobs'])
temp_df.head()
```

```python
#plotting the histogram
temp_df.plot.hist(bins=12)
plt.xlabel('Temperature (Bins=12)', fontsize=10)
plt.ylabel("Frequency", fontsize=10)
plt.title("Temperature Plot for Most Active Station ",fontsize=12)
plt.tight_layout()
plt.savefig('temperature_results_hist.png')
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates

#I have made the changes so that it retruns the details a year ago from the given dates.

def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    previous_start_date=dt.date(int(start_date.split("-")[0]),int(start_date.split("-")[1]),int(start_date.split("-")[2]))-dt.timedelta(days=365)
    previous_end_date=dt.date(int(end_date.split("-")[0]),int(end_date.split("-")[1]),int(end_date.split("-")[2]))-dt.timedelta(days=365)
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= previous_start_date).filter(Measurement.date <= previous_end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

trip=calc_temps('2012-02-28', '2012-03-05')
print(type(trip))
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)   
x=[1]
y=trip[0][1]
err=trip[0][0]-trip[0][2]

fig, ax = plt.subplots(figsize=(3,5))
plt.bar(x,y,width=0.1, yerr=err)
plt.xlim(0.9,1.1)
plt.ylim(0,trip[0][2]+10)
plt.title("Trip Avg Temp" )
plt.ylabel("Temp (F)")
```
